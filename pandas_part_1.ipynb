````python
# selecting columns

data = {
  "Name": ["Ram", "Shyam", "Ghanshyam", "Dhanshyam", "Aditi", "Jagdish", "Raj", "Geeta"],
  "Age": [28, 34, 22, 30, 29, 40, 25, 32],
  "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
  "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print("Sample Data Frame \n",df)

# Selecting a single column
print("Names( Single Column return series: ) \n")
# print(df["Name"]) or
names = df['Name']
print(names)

# selecting Multuple Columns
subset = df[["Name","Salary"]]
print("Subset of Data Frame \n", subset)
````

````python
#Filtering  Rows
data = {
  "Name": ["Ram", "Shyam", "Ghanshyam", "Dhanshyam", "Aditi", "Jagdish", "Raj", "Geeta"],
  "Age": [28, 34, 22, 30, 29, 40, 25, 32],
  "Salary": [50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
  "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print("Sample Data Frame \n",df)

##Single Condition Applying
high_salary = df[df['Salary'] > 50000]
print("\nEmployees with salary higher than 50000 : \n \n", high_salary)

##Multiple Condition Applying (AND)
filtered = df[(df['Age'] > 30) & (df['Salary'] > 50000)]
print("\nEmployees with age greater than 30 and salary higher than 50000 : \n \n", filtered)

##Multiple Condition Applying (OR )
filtered = df[(df['Age'] > 30) | (df['Salary'] > 50000)]
print("\nEmployees with age greater than 30 and salary higher than 50000 : \n \n", filtered)
````
