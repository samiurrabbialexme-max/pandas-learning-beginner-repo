## 1. Import pandas

`pandas` is imported as `pd`, which is the standard alias used in most data analysis projects.

```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("../data")
OUTPUT_DIR = Path("../outputs")
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 2. Read CSV file

`pd.read_csv()` reads a CSV file and converts it into a DataFrame.

```python
df = pd.read_csv(DATA_DIR / "sample_employees.csv")
df
```

## 3. Read JSON file

`pd.read_json()` reads JSON data. JSON is commonly used in APIs and web applications.

```python
json_df = pd.read_json(DATA_DIR / "sample_data.json")
json_df
```

## 4. Create a DataFrame manually

A Python dictionary can be converted into a pandas DataFrame. Keys become column names.

```python
data = {
    "Name": ["Sam", "Rabbi", "Alex"],
    "Age": [10, 20, 30],
    "City": ["Dhaka", "Chittagong", "Delhi"]
}

df_created = pd.DataFrame(data)
df_created
```

## 5. Save a DataFrame

`to_csv()` and `to_json()` export a DataFrame into files. `index=False` avoids saving an unnecessary index column in CSV.

```python
df_created.to_csv(OUTPUT_DIR / "output.csv", index=False)
df_created.to_json(OUTPUT_DIR / "output.json", orient="records", indent=2)

print("Files saved in outputs/ folder")
```

## 6. Dataset understanding with `head()` and `tail()`

`head()` shows the first rows. `tail()` shows the last rows.

```python
df = pd.read_csv(DATA_DIR / "sample_employees.csv")

df.head()
```

```python
df.tail()
```

## 7. Dataset information with `info()`

`info()` shows column names, non-null values, data types, and memory usage.

```python
df.info()
```

## 8. Descriptive statistics with `describe()`

`describe()` gives summary statistics for numeric columns.

```python
df.describe()
```

## 9. Shape and column names

`shape` returns rows and columns. `columns` returns column names.

```python
print("Shape:", df.shape)
print("Columns:", list(df.columns))
```

## 10. Select columns

Single-column selection returns a Series. Multiple-column selection returns a DataFrame.

```python
names = df["Name"]
names
```

```python
subset = df[["Name", "Salary"]]
subset
```

## 11. Filter rows

Boolean filtering is used to select rows that satisfy a condition.

```python
high_salary = df[df["Salary"] > 50000]
high_salary
```

## 12. Filter rows with AND condition

Use `&` when both conditions must be true. Always wrap each condition in parentheses.

```python
filtered_and = df[(df["Age"] > 30) & (df["Salary"] > 50000)]
filtered_and
```

## 13. Filter rows with OR condition

Use `|` when at least one condition must be true.

```python
filtered_or = df[(df["Age"] > 30) | (df["Salary"] > 50000)]
filtered_or
```
